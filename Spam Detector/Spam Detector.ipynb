```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
dvec = vectorizer.fit_transform(df['CONTENT'])
```

```python
dshuf = df.sample(frac=1)
```

```python
df_train = dshuf[:300]
df_test = dshuf[300:]
df_train_att = vectorizer.fit_transform(df_train['CONTENT'])  #fir bag-of-words on training set
df_test_att = vectorizer.transform(df_test['CONTENT'])
df_train_label = df_train['CLASS']
df_test_label = df_test['CLASS']
```

```python
df_train_att
```

```python
df_test_att
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=80)
```

```python
clf.fit(df_train_att, df_train_label)
```

```python
clf.score(df_test_att, df_test_label)
```

```python
from sklearn.metrics import confusion_matrix
pred_labels = clf.predict(df_test_att)
confusion_matrix(df_test_label, pred_labels)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, df_train_att, df_train_label, cv=5)

# show average score and +/- two standard deviations away (covering 95% of scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
# load all datasets and combine them
d = pd.concat([pd.read_csv("Youtube02-KatyPerry.csv"),
               pd.read_csv("Youtube03-LMFAO.csv"),
               pd.read_csv("Youtube04-Eminem.csv"),
               pd.read_csv("Youtube05-Shakira.csv")])
```

```python
len(d)
```

```python
len(d.query("CLASS == 1"))
```

```python
len(d.query("CLASS == 0"))
```

```python
dshuf = d.sample(frac=1)
d_content = dshuf['CONTENT']
d_label = dshuf['CLASS']
```

```python
# set a pipeline
from sklearn.pipeline import Pipeline, make_pipeline
pipeline = Pipeline([('bag-of-words',CountVectorizer()),
                    ('random-forest', RandomForestClassifier()),])
pipeline
```

```python
make_pipeline(CountVectorizer(), RandomForestClassifier())
```

```python
pipeline.fit(d_content[:1500],d_label[:1500])
```

```python
pipeline.score(d_content[1500:],d_label[1500:])
```

```python
pipeline.predict(["What a neat video!"])
```

```python
pipeline.predict(["plz subscribe to my channel"])
```

```python
scores = cross_val_score(pipeline, d_content, d_label, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
# add tfidf
from sklearn.feature_extraction.text import TfidfTransformer
pipeline2 = make_pipeline(CountVectorizer(),
                         TfidfTransformer(norm=None),
                         RandomForestClassifier())
sorted(pipeline.get_params().keys())
```

```python
scores = cross_val_score(pipeline2, d_content, d_label, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
pipeline2.steps
```

```python
#parameter search
parameters = {
    'countvectorizer__max_features': (None, 1000, 2000),
    'countvectorizer__ngram_range': ((1,1), (1,2)),
    'countvectorizer__stop_words': ('english', None),
    'tfidftransformer__use_idf': (True, False),
    'randomforestclassifier__n_estimators': (20, 50, 100)
}

from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(pipeline2, parameters, n_jobs=-1, verbose=1, cv=5)
```

```python
grid_search.fit(d_content, d_label)
```

```python
print("Best score: %0.3f" % grid_search.best_score_)
print("Best parameters set: ")
best_parameters = grid_search.best_estimator_.get_params()
for param_name in sorted(parameters.keys()):
    print("\t%s: %r" % (param_name, best_parameters[param_name]))
```
